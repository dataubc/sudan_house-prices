```python
from bs4 import BeautifulSoup

import requests


rooms_lists = []
bath_rooms_lists = []
price_lists = []
neighbourhood_lists= []
areas_lists = []

filecsv = open('SouqDataapple.csv', 'w', encoding='utf8')
for page in range(1,21):
    print('---', page, '---')
    soccer_requests = requests.get("https://sd.opensooq.com/en/real-estate-for-sale/apartments-for-sale?page="+str(page))

    soup = BeautifulSoup(soccer_requests.text,"html.parser")

    x = soup.find_all('div', {'class': 'rectLiDetails tableCell vMiddle pr15'})
    

    for i in range(len(x)):


        row= x[i]
        

        #print(z.find_all('a', {'class': 'block postLink'})), the address
        #print(z)
        try:
            bedrooms_li = row.find_all('li', {'class': 'ml8'}) 
            rooms_a = bedrooms_li[1].find_all('a')
            rooms = rooms_a[0]['title']
            rooms_lists.append(rooms)




            bath_rooms_a= bedrooms_li[2].find_all('a')
            bath_rooms = bath_rooms_a[0]['title'] 
            bath_rooms_lists.append(bath_rooms)

            areas_a= bedrooms_li[3].find_all('a')
            areas = areas_a[0]['title'] 
            areas_lists.append(areas)



            price_span =row.find_all('span', {'class': 'inline'})

            price =  price_span[1].get_text()
            price_lists.append(price)

            neighbourhood =  price_span[4].get_text()
            neighbourhood_lists.append(neighbourhood)
        
        except:
            pass
```

```python
print(rooms_lists)
print(bath_rooms_lists)
print(areas_lists)
print(neighbourhood_lists)
print(price_lists)
```

```python
import pandas as pd
list_all = [rooms_lists,bath_rooms_lists,areas_lists,neighbourhood_lists,price_lists]

df = pd.DataFrame(list_all, index=['bedrooms', 'bathrooms', 'area','neighbourhood','price']).T
```
